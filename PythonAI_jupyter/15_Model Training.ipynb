```python
import cv2
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from glob import glob
from pprint import pprint
from tqdm import tqdm
```

```python
LABEL_INT_DICT = np.unique(pd.read_csv('ucf11_train_vid.csv')['label'])
pprint(LABEL_INT_DICT) # 데이터의 타입과 형태 등도 같이 보여준다.(조금 더 예쁘게 출력해준다?)
LABEL_STR_DICT = {k:v for v,k in enumerate(LABEL_INT_DICT)}
pprint(LABEL_STR_DICT)
```

```python
import albumentations as A
import cv2
import math
import tensorflow as tf
from tensorflow import keras

class DataGenerator(keras.utils.Sequence):
    def __init__(self, batch_size, image_size, mode='train', shuffle=True):
        self.mode = mode
        self.shuffle = shuffle
        self.batch_size = batch_size
        self.image_size = image_size
        
        if self.mode == 'train':
            self.transform = A.Compose([
                A.HorizontalFlip(p=0.5),
                A.ShiftScaleRotate(
                    p=0.5,
                    shift_limit=0.05,
                    scale_limit=0.05,
                    rotate_limit=15,
                ),
            ])
            
        self.img_paths = glob(
            f'UCF11_updated_png/{mode}/*.png'
        )
        
    def __len__(self):
        return math.ceil(len(self.img_paths) / self.batch_size)
    
    def __getitem__(self, idx):
        strt = idx * self.batch_size
        fin = (idx + 1) * self.batch_size
        data = self.img_paths[strt:fin]
        
        batch_x, batch_y = self.get_data(data)
        return np.array(batch_x), np.array(batch_y)
    
    def get_data(self, data):
        batch_x = []
        batch_y = []
        
        for img_path in data:
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (self.image_size, self.image_size))
            
            if self.mode == 'train':
                augmented = self.transform(image=img)
                img = augmented['image']
            
            img = img.astype('float32')
            img = img / 255.
            
            label = img_path.split('\\')[-1].split('_')[0]
            label = LABEL_STR_DICT[label]
            
            batch_x.append(img)
            batch_y.append(label)
        return batch_x, batch_y
    
    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.img_paths)
```

```python
train_generator = DataGenerator(
    mode = 'train',
    batch_size = 128,
    image_size = 256,
    shuffle = True
)

valid_generator = DataGenerator(
    mode = 'valid',
    batch_size = 128,
    image_size = 256,
    shuffle = True
)
```

```python
for batch in train_generator:
    X, y = batch
    plt.figure(figsize = (15,15))
    
    for i in range(9):
        ax = plt.subplot(3,3,i+1)
        plt.imshow(X[i])
        plt.title(LABEL_INT_DICT[y[i]])
        plt.axis('off')
    break
    
```

```python
# tensorflow hub
# 
import tensorflow_hub as hub

model = tf.keras.Sequential([
    hub.KerasLayer("https://tfhub.dev/google/imagenet/efficientnet_v2_imagenet1k_b0/feature_vector/2",
                  trainable = True),
    tf.keras.layers.Dense(len(LABEL_INT_DICT),activation = 'softmax') # 수정한 부분/ 여러 개의 class: softmax
])

model.build([None,256,256,3]) # batch_size: 일단 생략, size = 256 * 256, color이니 3

adam = keras.optimizers.Adam(lr = 0.0001)
model.compile(
    optimizer = adam,
    loss = 'sparse_categorical_crossentropy', # categorical & one-hot-encoding 을 원하지 않음(수정해야 할 부분)
    metrics = 'accuracy'
)
model.summary()
```

```python
# 폴더 생성하기
import os
os.mkdir('ucf11_weights')
```

```python
filepath = 'ucf11_weights/{epoch:02d}-{value_accuracy:.2f}.hdf5'

model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath, monitor = 'val_loss',verbose = 1,save_best_only = True,
    save_weights_only = False, mode = 'min'
)
```

```python
history = model.fit(
    train_generator,
    validation_data = valid_generator,
    epochs = 10,
    verbose = 1,
    callbacks = [
        model_checkpoint
    ]
)
```

```python
import matplotlib.pyplot as plt
history = history.history

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(history['loss'], label='train')
plt.plot(history['val_loss'], label='val')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history['accuracy'], label='train')
plt.plot(history['val_accuracy'], label='val')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title("Accuracy")
plt.show()
```

```python
csv_path = 'ucf11_valid_vid.csv'
valid_df = pd.read_csv(csv_path)

value = 0

# tqdm: 실행 현황을 알려주는 라이브러리
for i, elem in tqdm(valid_df.iterrows(),total = len(valid_df)): # iterrows() : 행의 개수만큼 돌리기
    cap = cv2.VideoCapture(
        elem['file_path']
    )
    
    preds = []
    while True:
        ret,frame = cap.read()
        
        if not ret:
            break
        
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame,(256,256))
        frame = frame.astype('float32')
        frame /= 255.
        
        pred = model.predict(frame[np.newaxis,...]) # ... : 매개변수에 상관없이 대입하겠다
        preds.append(pred)
    
    preds = np.array(preds).mean(axis = 0) # 
    label = LABEL_INT_DICT[np.argmax(preds)] # 가장 많이 들어있는 값을 인덱스로 선택 -> label로 사용하겠다.
    
    if label == elem['label']:
        value += 1
    
    cap.release()

print("Accuracy: ",value / len(valid_df))
```
