### **1-1. Tensor**

* 텐서(tensor)는 배열(array)이나 행렬(matrix)과 매우 유사한, 특수한 자료구조
* Pytorch에서는 텐서를 사용하여 모델의 입력(input)과 출력(output), 그리고 모델의 매개변수들을 부호화(encode)함

```python
data = [[1,2],[3,4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(data)
x_np_1 = torch.tensor(np_array) # 별도의 복사본(새로운 tensor) 생성 -> 메모리의 낭비가 발생
print(x_np_1)
```

```python
x_np_1[0,0] = 10
print(x_np_1)
print(np_array)
```

```python
x_np_2 = torch.as_tensor(np_array) # 뷰를 만듦 -> 메모리의 낭비가 x(동일한 메모리를 사용하고 있다.)
print(x_np_2)
```

```python
x_np_2[0,0] = 10
print(x_np_2)
print(np_array)
```

```python
x_np_3 = torch.from_numpy(np_array) # 뷰를 만듬
print(x_np_3) 
```

```python
x_np_3[0,0] = 10
print(x_np_3)
print(np_array)
```

```python
# 텐서 -> ndArray
np_again = x_np_1.numpy()
print(np_again,type(np_again))
```

### **1-2. Tensor의 속성**


```python
tensor = torch.rand(3,4)

print(f'Shape of tensor : {tensor.shape}')
print(f'Datatype of tensor : {tensor.dtype}')
print(f'Device of tensor : {tensor.device}')
```

```python
# 속성 변경
tensor = tensor.reshape(4,3)
tensor = tensor.int()
if torch.cuda.is_available(): # GPU를 사용하는 것이 가능하니??
  tensor = tensor.to('cuda') # GPU로 바꿔줘!!

print(f'Shape of tensor : {tensor.shape}')
print(f'Datatype of tensor : {tensor.dtype}')
print(f'Device of tensor : {tensor.device}')
```

### **1-4.Transpose**

```python
a = torch.arange(16).reshape(2,2,4)
print(a,a.shape)
```

```python
b = a.transpose(1,2)
print(b,b.shape)
```

```python
c = a.permute(2,0,1)
print(c,c.shape) # transpose와 동일한 기능
```

### **1-5. Tensor 연산**

```python
x = torch.tensor([[1,2],[3,4]],dtype = torch.float32)
y = torch.tensor([[5,6],[7,8]],dtype = torch.float32)
print(x)
print(y)
```

```python
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x@y) 
print("*" *30)
print(torch.add(x,y))
print(torch.subtract(x,y))
print(torch.multiply(x,y))
print(torch.divide(x,y))
print(torch.matmul(x,y))
```

```python
# in - place 연산(복합대입연산자 느낌?)
print(x.add(y))
print(x)

print(x.add_(y)) # 기존의 x에 y를 더한 값을 다시 x에 저장하라~
print(x)
# 나머지도 비슷함
```

```python
z = torch.arange(1,11).reshape(2,5)
print(z)
```

```python
sum1 = torch.sum(z,axis = 0)
sum2 = torch.sum(z,axis = 1)
sum3 = torch.sum(z,axis = -1)
print(sum1,sum1.shape)
print(sum2,sum2.shape)
print(sum3,sum3.shape)
```

```python
a = torch.arange(24).reshape(4,6)
b = a.clone().detach() # 복사된 내용(ctrl+c)을 b에 붙여서 저장해라~(ctrl+v)
print(a,a.shape)
print(b,b.shape)
```

```python
c = torch.cat([a,b],axis = -1)
print(c,c.shape)
```
