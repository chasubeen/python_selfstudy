### **1. 표준편차**


* 자료의 산포도를 나타내는 수치로, 분산의 양의 제곱근으로 정의
* 표준편차가 작을수록 평균값에서 변량들의 거리가 가까움
* $$\sqrt{\frac{\sum_{k=1}^N (x_i - \bar{x})^2}{n}}$$


```python
# 표준편차를 계산할 데이터의 개수를 입력하고, 입력받은 개수만큼 데이터를 입력

# 표준편차를 계산할 데이터의 개수를 입력하세요: 

# 1번쨰 데이터:
# 2번쨰 데이터: 
# ...
# 5번째 데이터:

n = int(input("표준편차를 계산할 데이터의 개수를 입력하세요: "))

data = []
for i in range(n):
  a = int(input(str(i+1)+"번째 데이터: "))
  data.append(a)

print(data)
```

```python
# 강사님 코드
n = int(input("표준편차를 계산할 데이터의 개수를 입력하세요: "))

score = [0 for i in range(n)]  # n개만큼 0으로 초기화됨
print(score)

for i in range(n):
  data = int(input('{}번째 데이터: '.format(i+1)))
  score[i] = data

print(score)
```

```python
# 내 코드
# score의 최댓값과 최솟값 구하기 

# print(score)
# 최댓값: , 최솟값: 
print(score)

max_num = score[0]
min_num = score[0]

for i in range(n): 
  if score[i] > max_num:
    max_num = score[i]
  if score[i] < min_num:
    min_num = score[i]

print("최댓값: {}".format(max_num),end=' ')
print("최솟값: {}".format(min_num),end=' ')
```

```python
# 강사님 코드
for i in range(len(score)-1):
  for j in range(i+1,len(score)):
    if score[i] > score[j]:
      score[i],score[j] = score[j],score[i] # 정렬하기
print(score)
print("최소값: {}, 최대값: {}". format(score[0],score[len(score)-1]))
```

```python
# 최댓값과 최솟값을 제외한 나머지 데이터의 평균을 계산
# 내 코드
new_score = []
for i in score:
  if (i != max_num) and (i != min_num):
    new_score.append(i)

print(new_score)
# mean:
sum = 0
for i in new_score:
  sum += i

avg = sum / len(new_score)
print(avg)
```

```python
# 강사님 코드
total = 0
for i in range(1,len(score)-1):
  total += score[i]

mean = total / (n-2)
print(mean)
```

```python
# 분산
# ∑Nk=1(xi−x¯)2

# 내 코드
sum2 = 0
for i in score:
  sum2 += i
avg2 = sum2 / len(score)
print(avg2)

sum3 = 0
for j in score:
  sum2 += (j-avg2)**2

var = sum3 / len(score)
print(var)
```

```python
total = 0.0
for i in range(1,len(score)-1):
  total += (score[i] - mean) ** 2

var = total / (n-2)
print(var)

import math
std = math.sqrt(var) # sqrt() : 제곱근
print(std)
```

```python
print('{0}에서 최댓값 {1}, 최솟값 {2}를 제외한 데이터의 평균은 {3:6.4f}이고, 분산은 {4:6.4f}이고, 표준편차는 {5:6.4f}입니다.'.format(score,score[n-1],score[0],mean,var,std))
```
