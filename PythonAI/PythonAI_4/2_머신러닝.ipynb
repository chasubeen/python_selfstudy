### **1. 머신러닝이란?**

* 인공지능(AI): 사람의 지능을 모방하여 사람이 하는 것과 같이 복잡한 일을 할 수 있게 기계를 만드는 것
* 머신러닝: 기본적으로 알고리즘을 이용해 데이터를 분석 및 학습하여, 학습한 내용을 기반으로 판단이나 예측을 하는 것
* 딥러닝: 인공신경망에서 발전한 형태의 인공지능. 머신러닝 중 하나의 방법론

> 머신러닝은 데이터(Data)를 기반으로 패턴(Model)을 학습하여 결과(Prediction)을 추론

##### **1-1.학습**
* 지도학습:회귀, 분류(정답 데이터가 있는 학습)
* 비지도학습:군집화(컴퓨터가 스스로 분류), 차원축소(정답 데이터가 없는 학습)
* 강화 학습

##### **1-2. 머신러닝의 장점**


* 복잡한 패턴을 인지
* 적절한 알고리즘, 다양한 양질의 데이터가 있다면 좋은 성능
* 도메인 영역(전문분야)에 대한 지식이 상대적으로 부족해도 가능

##### **1-3. 머신러닝의 문제점**

* 데이터의 의존성이 큼
* 과적합의 오류(일반화의 오류, 데이터 다양성 요구)
* 풀부한 데이터가 기본적으로 요구됨
> 좋은 성능 = 좋은 데이터(quality, quantity)
