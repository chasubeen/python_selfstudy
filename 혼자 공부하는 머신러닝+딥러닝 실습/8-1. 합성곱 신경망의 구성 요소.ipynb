###**합성곱**
- 입력 데이터에서 유용한 특성만을 드러나게 하는 딥러닝 기법
- 특정 개수의 가중치를 가지고 동일한 출력을 반복함
- 계산 결과 차원이 '유지'됨
- 커널: 입력에 곱하는 가중치/ 주로 (3,3)이나 (5,5) 크기가 권장됨
- 필터: 뉴런의 개수
- 특성 맵(feature map): 합성곱 계산을 통해 얻은 출력/ (너비,높이,차원)

### **케라스 합성곱 층**

```python
# ConvSD 클래스: 입력 위를 이동(왼 -> 오/ 위 -> 아래)하는 합성곱 클래스
from tensorflow import keras
keras.layers.Conv2D(10,kernel_size = (3,3),activation = 'relu') # 필터의 개수, 필터에 사용할 커널의 크기, 활성화 함수
```

###**합성곱 신경망**
- 1개 이상의 합성곱 층을 쓴 인공 신경망


###**패딩과 스트라이드**
- 패딩(padding): 입력 배열의 주위를 가상의 원소로 채우는 것
- 실제 입력값이 아니기 때문에 패딩은 0으로 채운다.
- same padding: 입력 == 특성 맵
- 스트라이드(stride): 커널 이동의 크기

```python
keras.layers.Conv2D(10, kernel_size =  (3,3),activation = 'relu',padding = 'same',strides = 1)
```

###**풀링**
- 합성곱 층에서 만든 특성 맵의 가로/세로 크기를 줄이는 역할
- 특성 맵의 개수는 줄지 x
- 가중치가 x -> but 최댓값이나 평균값을 계산하는 역할을 수행
- 풀링 영역은 겹치지 않게 이동하는 특성이 있음
