```python
# 단일 그래프 그리기
data = np.arange(1,100) # 1 이상 100 미만
plt.plot(data) # 단일 그래프에 대한 '설정'
# 그래프를 보여주는 코드
plt.show()
```

```python
# figure()는 새로운 그래프 canvas를 생성
data = np.arange(100,201)
plt.plot(data)

data2 = np.arange(200,301)
plt.figure()
plt.plot(data2)

plt.show()
```

```python
# 여러 개의 plot을 그리는 방법
# subplot(row,column,index)
# 각 그래프 canvas가 행렬 구조로 출력된다.
data = np.arange(100,201)
plt.subplot(2,1,1)  # 2행 1열의 데이터로 그래프가 출력된다.
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(2,1,2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100,201)
plt.subplot(211)  # 콤마를 생략하고 속성을 부여해도 된다.(비추)
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
# 문제
# 1행 3열짜리 그래프 3개 작성하기
# 1행 1열(100~200), 1행 2열(200~300) , 1행 3열(300~400)

# 내 코드(강사님과 동일)
data = np.arange(100,201)
plt.subplot(1,3,1) 
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(1,3,2)
plt.plot(data2)

data3 = np.arange(300,401)
plt.subplot(1,3,3)
plt.plot(data3)

plt.show()
```

```python
# plt.subplots(행의 개수, 열의 개수)
data = np.arange(1,51)

fig,axes = plt.subplots(2,3)

axes[0,0].plot(data)
axes[0,1].plot(data*data)
axes[0,2].plot(data**3)
axes[1,0].plot(data%10)
axes[1,1].plot(-data)
axes[1,2].plot(data // 20)

plt.tight_layout() 
plt.show()
```

#### **4-7. 막대 그래프 그리기**
**4-7-1. Barplot**  
- plt.bar(x,y) --> 일반적인 막대 그래프(세로방향)
- align = 'center' --> 막대 그래프를 '가운데'에 정렬(default값)  

**4-7-2. Barhplot**  
- plt.barh(x,y)
- barh 함수는 막대그래프의 방향을 '가로'로 표시

```python
x = ['Math','Programming','Data Science', 'Art','English','Physics']
y = [66,80,60,50,80,10]

plt.figure(figsize = (6,3))
# plt.bar(x,y)
plt.bar(x,y,align = 'center',alpha = 0.7,color = 'red')    # align = 'center' : 가운데 정렬(default 값)
plt.xticks(x)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
# barh 함수에서는 xticks으로 설정했던 부분을 yticks로 변경(방향 바뀌었으니..)

# Barplot, Barhplot 
x = ['Math','Programming','Data Science', 'Art','English','Physics']
y = [66,80,60,50,80,10]

plt.figure(figsize = (6,3))
# plt.bar(x,y)
plt.barh(x,y,align = 'center',alpha = 0.7,color = 'green')    # align = 'center' : 가운데 정렬(default 값)
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
x_label = ['Math','Programming','Data Science', 'Art','English','Physics']
x = np.arange(len(x_label))
print(x)
y_1 = [66,80,60,50,80,10]
y_2 = [55,90,40,60,70,20]

width = 0.35

fig,axes = plt.subplots()

axes.bar(x - width/2,y_1,width,align = 'center',alpha = 0.5)
axes.bar(x + width/2,y_2,width,align = 'center',alpha = 0.8)

plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title("Subjects")

plt.legend(['김사과','반하나'])

plt.show()
```

```python
# 문제
# 회전, 김사과의 점수는 초록색으로 표현, 반하나의 점수는 빨간색으로 표현하기

x_label = ['Math','Programming','Data Science', 'Art','English','Physics']
x = np.arange(len(x_label))
print(x)
y_1 = [66,80,60,50,80,10]
y_2 = [55,90,40,60,70,20]

width = 0.35

fig,axes = plt.subplots()

axes.barh(x - width/2,y_1,width,align = 'center',alpha = 0.5, color = 'green')
axes.barh(x + width/2,y_2,width,align = 'center',alpha = 0.8, color = 'red')

plt.yticks(x)
axes.set_yticklabels(x_label)
plt.xlabel('Number of Students')
plt.title("Subjects")

plt.legend(['김사과','반하나'])

plt.show()
```

#### **4-8. 선 그래프 그리기**

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x,y_1,label = '1 + sin',color = 'blue', alpha = 0.3)
plt.plot(x,y_2,label = '1 + cos',color = 'red', alpha = 0.7)

plt.xlabel('x value',fontsize = 15)
plt.ylabel('y value',fontsize = 15)
plt.title('sin and cos graph',fontsize = 20)

plt.grid()
plt.legend()

plt.show()
```

#### **4-10.Pie Chart**

```python
# Pie Chart
labels = ['삼성', '카카오', '로블룩스', '애플', '메타', 'AT&T']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]  # 각 영역의 사이즈
explode = (0.3, 0, 0.1, 0, 0, 0) # 각 영역이 앞으로 튀어나온 정도

patches, texts, autotexts = plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)

plt.title('성공의 지름길', fontsize=20)

# label 텍스트에 대한 스타일 적용
for t in texts:
  t.set_fontsize(12)
  t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
  t.set_color('white')
  t.set_fontsize(17)

plt.show()
```

#### **4-11. imshow**
- 이미지(image) 데이터와 유사하게 행과 열을 가진 2차원의 데이터를 시각화

```python
from sklearn.datasets import load_digits

digits = load_digits() # 0~16의 값을 가지는 array로 구성되어 있음
X = digits.images[:10] 
X[0] # 1개의 array는 8*8 배열 안에 표현되어 있음, 숫자는 0~9로 표현
```

```python
# 8시 44분경 녹화본 돌려보기(주석 달기)
fig, axes = plt.subplots(nrows = 2, ncols = 5,sharex = True,figsize = (12,6),sharey = True)
# 2행 5열로 canvas 배열, x와 y의 값은 모두 동일하게

for i in range(10):
  axes[i//5][i%5].imshow(X[i],cmap = 'Blues')
  axes[i//5][i%5].set_title(str(i),fontsize = 20)

plt.tight_layout()

plt.show()
```
