# **1. 손실 함수를 사용하는 목표**

* Linear Layer 함수가 얼마나 원하는 만큼 잘 동작하는지 측정하고 싶음
* 동작을 점수로 나타내고 싶음

# **2. Loss**

* 손실값(Loss): 원하는 출력값(target,y)과 실제 출력값(output,y^)의 차이의 합
* Loss가 작을수록 가상의 linear함수를 잘 묘사하고 있다고 할 수 있음
* Loss가 작은 Linear Layer를 선택하면 된다.

# **3. Loss Fuction**


* Linear Layer의 parameter를 바꿀 때마다 Loss를 계산

> 목표로 하는 함수를 묘사하기 위해 학습용 입력 데이터들을 Linear Layer에 넣어 출력값들을 구하고, 출력값(y^)들과 목표값(y)들의 차이의 합(Loss)을 최소화 해야 함


# **4. MSE(Mean Square Error) Loss**


$$
\text{MSE}(\hat{x}_{1:N}, x_{1:N})=\frac{1}{N\times{n}}\sum_{i=1}^{N}{||x_i-\hat{x}_i||_2^2}
$$


```python
import torch
```

```python
def mse(x_hat,x):
  result = ((x_hat - x) ** 2).mean()
  return result
```

```python
x = torch.FloatTensor([[1,1],
                       [2,2]])
x_hat = torch.FloatTensor([[0,0],
                           [0,0]])

print(x.size(),x_hat.size())
```

```python
mse(x_hat,x)
```

# **5. Pytorch 내의 손실함수 사용하기**

```python
import torch.nn.functional as F
```

```python
F.mse_loss(x_hat,x,reduction = 'sum')
```

```python
F.mse_loss(x_hat,x,reduction = 'none') # (x-x_hat) ** 2
```

```python
import torch.nn as nn

mse_loss = nn.MSELoss()
mse_loss(x_hat,x)
```
