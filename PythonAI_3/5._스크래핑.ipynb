### **2-1. basicenglishspeaking.com 데이터 크롤링**

```python
# 크롤링을 위한 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site) # 사이트에 접속하겠다.
# print(request)  # <Response [200]>  : 정상적인 응답을 받았다.
html = request.text # 접속한 사이트에서 문자에 해당되는 정보만을 html 변수에 저장해라.
# print(html)
soup = BeautifulSoup(html,'html.parser')  # html.parser: html을 분석하겠다.
# print(soup)
```

```python
subject = []
divs = soup.findAll('div',{'class':'thrv-columns'}) # div 태그의 class가 'thrv-columns'인 것들만을 모두 찾겠다.
print(divs)
```

```python
for div in divs:
  links = div.findAll("a") # divs에서 a를 모두 찾아(앵커 태그를 찾아) links에 저장해라.
  for link in links:
    # print(link.text)
    subject.append(link.text) # link에서 텍스트만을 추출하여(태그를 모두 제거하고) 저장하라.
```

```python
subject
```

```python
print('총',len(subject),'개의 대화를 찾았습니다.')
for i in range(len(subject)):
  print('{0:2d},{1:s}'.format(i+1,subject[i]))
```

### **2-2. 벅스 뮤직 차트 크롤링** 
- https://music.bugs.co.kr/chart

```python
request = requests.get('https://music.bugs.co.kr/chart')
html = request.text
soup = BeautifulSoup(html,'html.parser')

titles = soup.findAll('p',{'class':'title'})
# print(titles)

artists = soup.findAll('p',{'class':'artist'})
# print(artists)
```

```python
for i in range(len(titles)):
  title = titles[i].text.strip()
  # print(title)
  artist = artists[i].text.strip().split('\n')[0]
  # print(artist)
  data = '{0:3d}위 {1} -{2}'.format(i+1,artist,title)
  print(data)
```

### **2-4. 네이트 판 크롤링하기**
- 베스트 댓글, 일반 댓글

```python
from urllib.request import urlopen
```

```python
url = 'https://pann.nate.com/talk/364288844'
soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
comment_best = soup.find('div', class_='cmt_best') # 영역 지정하기(하나만 찾아오기)
comment_best = comment_best.find_all('dd', class_='usertxt') # 싹 다 찾아오기
# print(comment_best)
print('👍 베스트 댓글')
for comment in comment_best:
  print(comment.get_text().strip())
  print('*'*50)
```

```python
# 과제
# 일반 댓글 뽑아보기
url = 'https://pann.nate.com/talk/364288844'
soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
comment_general = soup.find('div',class_ = 'cmt_list')
comment_general = comment_general.find_all('dd',class_ = 'usertxt')
# print(comment_general)
print("")
for comment in comment_general:
  print(comment.get_text().strip())
  print("="*50)
```

```python
# 강사님 코드
comment_list = soup.find('div', class_='cmt_list') 
comment_list = comment_list.find_all('dd',class_ = "usertxt")
# print(comment_list)
print("😊 일반 댓글")
for comment in comment_list:
  print(comment.get_text().strip())
  print("="*50)
```

### **2-5. 다음 뉴스 기사 제목 크롤링**

```python
# https://news.v.daum.net/v/20220103175157964
# https://news.v.daum.net/v/20220103184616242

def get_daum_news_title(news_id):
  url = 'https://news.v.daum.net/v/'+news_id
  soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
  title = soup.find('h3',class_ = 'tit_view')
  print(title.get_text().strip())
```

```python
# 강사님 코드
def get_daum_news_title(news_id):
  url = 'https://news.v.daum.net/v/{}'.format(news_id)
  request = requests.get(url)
  soup = BeautifulSoup(request.text)
  title = soup.select_one('h3.tit_view')  # h3 태그의 tit_view 클래스)
  if title:
    return title.get_text()
  else:
    return ""  # 빈 문자열을 리턴한다.
```
